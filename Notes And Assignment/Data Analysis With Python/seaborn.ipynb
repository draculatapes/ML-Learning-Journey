# Data Visulation With Seaborn

seaborn is a Python visualization library based on Matplotlib that provided a high-level interface for drawing attractive and informative statistical graphics . Seaborn helps in creating complex visulization with just a few lines of code. In this lesson , we will cover the basics of seaborn, including creating various types of plots and cusomizing them.

```python
import seaborn as sns
```

```python
## Basic Plotting with Seaborn 

tips=sns.load_dataset('tips')
```

```python
print(type(tips))
tips
```

```python
### ceraate a scatterplot

import matplotlib.pyplot as plt
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title('scatterplot of total bill vs tip ')
```

```python
## line Plot
sns.lineplot(x='size',y='total_bill',data=tips)
plt.title('line plot to total bill by size ')
plt.show()
```

```python
### Categorical Plots

## Bar Plot
sns.barplot(x='day', y='total_bill',data=tips)
plt.title('bar plot of total ')
plt.show()
```

```python
## Box Plot

sns.boxplot(x='day',y='total_bill',data=tips)
```

```python

### Violin Plot

sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
## histograms 

sns.histplot(tips['total_bill'],bins=10,kde=True)
```

```python
## Kde plot

sns.kdeplot(tips['total_bill'],fill=True)
```

```python
### PairPlot

sns.pairplot(tips)
```

```python
## HeatMap

corr=tips[['total_bill','tip','size']].corr()
corr
```

```python
sns.heatmap(corr,annot=True,cmap='coolwarm')
```
