Pandas provides a wide range of functions for data manipulation and analysis, making it easier to clean , transform and extract insights from data . In this lesson , we will cover various data manipulation and anaylysis techniques using pandas and numpy

```python
## Handling Missing Values 

df.isnull().any()
```

```python
df.isnull().sum()
```

```python
df_filled=df.fillna(0)
```

```python
## Filling missing values with the mean fo the column
df['Sales_fillNA']=df['Sales'].fillna(df['Sales'].mean())
```

```python
df
```

```python
df.dtypes
```

```python
## Renaming Columns
df=df.rename(columns={'Date':'Sale Date'})
df
```

```python
## change datatypes
df['Value_new']=df['Value'].fillna(df['Value'].mean()).astype(int)
```

```python
df.dtypes
```

```python
df['New Value']=df['Value'].apply(lambda x: x*2)
```

```python
df
```

```python
## Data Aggregation and Grouping 

df.head(10)
```

```python
grouped_mean=df.groupby('Product')['Value'].mean()
print(grouped_mean)
```

```python
series=df.groupby('Product')
print(series['Value'].std())
```

```python
group_sum=df.groupby(['Product','Region'])['Value'].sum()
print(group_sum)
```

```python
## aggregate multiple functions 
grouped_agg=df.groupby('Region')['Value'].agg(['mean','sum','count'])
grouped_agg
```

```python
### Merging and Joining Data Frames
df1=pd.DataFrame({'key':['A','B','C'],'Value1':[1,2,3]})
df2=pd.DataFrame({'key': ['A','B','D'],'Value2': [4,5,6]})
print(df1.columns)
print(df2.columns)
```

```python
## Merge DataFrame on the 'Key Columns'
pd.merge(df1,df2,on="key",how="inner")
```

```python
pd.merge(df1,df2,on="key",how="outer")
```

```python
pd.merge(df1,df2,on="key",how="left")
```

```python
pd.merge(df1,df2,on="key",how="right")
```
